# 📦 Supply Chain Inventory & Logistics Analysis
**Dataset:** DataCo Supply Chain Dataset (180,519 orders)  
**Tools:** Python (pandas, matplotlib, seaborn)  
**Goal:** Clean, explore and analyse supply chain data, then export a clean file for Power BI dashboarding.

---

## Step 1 — Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set_theme(style='whitegrid')
```

## Step 2 — Load the Dataset

```python
df = pd.read_csv(
    r"C:\Users\User\Downloads\archive (12)\DataCoSupplyChainDataset.csv",
    encoding='latin-1'
)

print('Shape:', df.shape)
print('\nColumn Names:')
print(df.columns.tolist())
df.head()
```

## Step 3 — Explore the Data

```python
df.info()
```

```python
# Check missing values
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
print('Columns with missing values:')
print(missing_df[missing_df['Missing Count'] > 0])
```

```python
# Basic statistics on key numeric columns
df[['Sales', 'Order Item Profit Ratio', 'Benefit per order',
    'Days for shipping (real)', 'Days for shipment (scheduled)',
    'Order Item Discount Rate', 'Order Item Quantity']].describe().round(2)
```

## Step 4 — Clean the Data

```python
# Drop PII and irrelevant columns (safely)
cols_to_drop = [
    'Customer Email', 'Customer Password', 'Customer Street',
    'Customer Zipcode', 'Product Description', 'Product Image',
    'Order Zipcode'
]
cols_to_drop = [col for col in cols_to_drop if col in df.columns]
df.drop(columns=cols_to_drop, inplace=True)
print(f'Dropped {len(cols_to_drop)} unnecessary columns')

# Drop duplicates
before = len(df)
df.drop_duplicates(inplace=True)
print(f'Duplicates removed: {before - len(df)}')

# Parse date columns
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'], errors='coerce')
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'], errors='coerce')

# Extract time features
df['Order Year'] = df['order date (DateOrders)'].dt.year
df['Order Month'] = df['order date (DateOrders)'].dt.month
df['Order Month Name'] = df['order date (DateOrders)'].dt.strftime('%b')

print('Dates parsed')
print('Final shape:', df.shape)
```

## Step 5 — Feature Engineering (Creating KPIs)

```python
# Shipping delay in days (positive = late)
df['Shipping Delay (days)'] = (
    df['Days for shipping (real)'] - df['Days for shipment (scheduled)']
)

# Human-readable late flag
df['Is Late'] = df['Late_delivery_risk'].map({1: 'Late', 0: 'On Time'})

# Discount amount in dollars
df['Discount Amount'] = (
    df['Order Item Product Price'] * df['Order Item Discount Rate']
).round(2)

# Net revenue after discount
df['Net Revenue'] = (df['Sales'] - df['Discount Amount']).round(2)

df[['Order Id', 'Product Name', 'Sales', 'Net Revenue',
    'Shipping Delay (days)', 'Is Late', 'Benefit per order']].head(10)
```

## Step 6 — Analysis & Visualisations

### 6a. Top 10 Categories by Total Sales

```python
sales_by_cat = (
    df.groupby('Category Name')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10, 6))
sales_by_cat.plot(kind='bar', color='steelblue', edgecolor='white')
plt.title('Top 10 Categories by Total Sales', fontsize=14, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('sales_by_category.png', dpi=150)
plt.show()
```

### 6b. Late Delivery Rate by Shipping Mode

```python
late_by_mode = (
    df.groupby('Shipping Mode')['Late_delivery_risk']
    .mean()
    .mul(100)
    .round(2)
    .sort_values(ascending=False)
)

plt.figure(figsize=(8, 5))
late_by_mode.plot(kind='bar', color='tomato', edgecolor='white')
plt.title('Late Delivery Rate (%) by Shipping Mode', fontsize=14, fontweight='bold')
plt.xlabel('Shipping Mode')
plt.ylabel('Late Delivery Rate (%)')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig('late_delivery_by_mode.png', dpi=150)
plt.show()
print(late_by_mode)
```

### 6c. Monthly Sales Trend

```python
monthly_sales = (
    df.groupby(['Order Year', 'Order Month'])['Sales']
    .sum()
    .reset_index()
)
monthly_sales['Period'] = (
    monthly_sales['Order Year'].astype(str) + '-' +
    monthly_sales['Order Month'].astype(str).str.zfill(2)
)
monthly_sales = monthly_sales.sort_values('Period')

plt.figure(figsize=(14, 5))
plt.plot(
    monthly_sales['Period'], monthly_sales['Sales'],
    marker='o', color='steelblue', linewidth=2
)
plt.title('Monthly Sales Trend', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('monthly_sales_trend.png', dpi=150)
plt.show()
```

### 6d. Profit by Customer Segment

```python
profit_by_segment = (
    df.groupby('Customer Segment')['Benefit per order']
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(7, 5))
profit_by_segment.plot(kind='bar', color='mediumseagreen', edgecolor='white')
plt.title('Total Profit by Customer Segment', fontsize=14, fontweight='bold')
plt.xlabel('Customer Segment')
plt.ylabel('Total Profit ($)')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig('profit_by_segment.png', dpi=150)
plt.show()
print(profit_by_segment)
```

### 6e. Net Revenue by Market

```python
revenue_by_market = (
    df.groupby('Market')['Net Revenue']
    .sum()
    .sort_values(ascending=True)
)

plt.figure(figsize=(8, 5))
revenue_by_market.plot(kind='barh', color='#2196F3', edgecolor='white')
plt.title('Net Revenue by Market', fontsize=14, fontweight='bold')
plt.xlabel('Net Revenue ($)')
plt.ylabel('Market')
plt.tight_layout()
plt.savefig('revenue_by_market.png', dpi=150)
plt.show()
print(revenue_by_market)
```

### 6f. Delivery Status Distribution

```python
delivery_counts = df['Delivery Status'].value_counts()

plt.figure(figsize=(7, 5))
delivery_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['#4CAF50', '#F44336', '#FF9800', '#2196F3'],
    startangle=90,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)
plt.title('Delivery Status Distribution', fontsize=14, fontweight='bold')
plt.ylabel('')
plt.tight_layout()
plt.savefig('delivery_status.png', dpi=150)
plt.show()
print(delivery_counts)
```

### 6g. Correlation Heatmap

```python
numeric_cols = [
    'Sales', 'Net Revenue', 'Benefit per order',
    'Order Item Quantity', 'Order Item Discount Rate',
    'Days for shipping (real)', 'Shipping Delay (days)',
    'Late_delivery_risk', 'Order Item Profit Ratio'
]

plt.figure(figsize=(11, 7))
sns.heatmap(
    df[numeric_cols].corr().round(2),
    annot=True,
    cmap='coolwarm',
    linewidths=0.5,
    fmt='.2f'
)
plt.title('Correlation Heatmap of Key Metrics', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=150)
plt.show()
```

## Step 7 — Summary Insights

```python
print('=' * 55)
print('KEY INSIGHTS SUMMARY')
print('=' * 55)

print(f"\nTotal Sales:              ${df['Sales'].sum():,.2f}")
print(f"Total Net Revenue:        ${df['Net Revenue'].sum():,.2f}")
print(f"Total Profit:             ${df['Benefit per order'].sum():,.2f}")
print(f"Total Orders:             {df['Order Id'].nunique():,}")
print(f"Late Delivery Rate:       {df['Late_delivery_risk'].mean()*100:.1f}%")
print(f"Avg Shipping Delay:       {df['Shipping Delay (days)'].mean():.1f} days")
print(f"Avg Discount Rate:        {df['Order Item Discount Rate'].mean()*100:.1f}%")

print(f"\nTop Category by Sales:    {df.groupby('Category Name')['Sales'].sum().idxmax()}")
print(f"Top Market by Revenue:    {df.groupby('Market')['Net Revenue'].sum().idxmax()}")
print(f"Top Customer Segment:     {df.groupby('Customer Segment')['Benefit per order'].sum().idxmax()}")
print(f"Riskiest Shipping Mode:   {df.groupby('Shipping Mode')['Late_delivery_risk'].mean().idxmax()}")
```

## Step 8 — Export Cleaned Data for Power BI

```python
# Select the most useful columns for Power BI
powerbi_cols = [
    'Order Id', 'order date (DateOrders)', 'Order Year', 'Order Month', 'Order Month Name',
    'Customer Segment', 'Customer City', 'Customer Country',
    'Market', 'Order Region', 'Order Country', 'Order City',
    'Category Name', 'Department Name', 'Product Name', 'Product Price',
    'Order Item Quantity', 'Sales', 'Net Revenue', 'Benefit per order',
    'Order Item Discount Rate', 'Discount Amount',
    'Shipping Mode', 'Delivery Status', 'Is Late',
    'Days for shipping (real)', 'Days for shipment (scheduled)', 'Shipping Delay (days)',
    'Late_delivery_risk', 'Order Item Profit Ratio', 'Order Status', 'Type'
]

df_powerbi = df[powerbi_cols].copy()
df_powerbi.to_csv('supply_chain_cleaned.csv', index=False, encoding='utf-8')

print('Cleaned dataset exported as supply_chain_cleaned.csv')
print(f'Rows: {len(df_powerbi):,} | Columns: {len(df_powerbi.columns)}')
```
